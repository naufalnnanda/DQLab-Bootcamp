#Data Cleaning

Menghapus nilai yang tidak sesuai pada Kolom RR
pada keterangan Data, nilai RR = 8888 artinya data tidak terukur, dan RR = 9999 artinya tidak dilakukan pengukuran


```python
# Mengganti nilai 9999 (dianggap tidak sesuai) dengan NaN
df['RR'] = df['RR'].replace(9999, 0)
```

```python
df.describe()
```

```python
# Mengganti nilai 8888 (dianggap tidak sesuai) dengan NaN
df['RR'] = df['RR'].replace(8888, np.nan)
```

```python
df.describe()
```

Mengganti nilai null dengan data di baris atas atau bawahnya dengan menggunakan interpolasi

```python
# Mengganti nilai null dengan interpolasi linier
df['RR'] = df['RR'].interpolate(method='linear')
```

```python
df.isnull().sum()
```

```python
df['TN'] = df['TN'].interpolate(method='linear')
df['TX'] = df['TX'].interpolate(method='linear')
df['TAVG'] = df['TAVG'].interpolate(method='linear')
df['RH_AVG'] = df['RH_AVG'].interpolate(method='linear')
df['SS'] = df['SS'].interpolate(method='linear')
df['DDD_X'] = df['DDD_X'].interpolate(method='linear')
```

Mengubah nilai derajat pada kolom DDD_X (Arah Angin pada Kecepatan Maksimum) ke dalam karakter Arah Mata Angin

```python
def derajat_ke_arah_pandas(derajat_series):
    bins = [0, 23, 68, 113, 158, 203, 248, 293, 338, 360]
    labels = ["N", "NE", "E", "SE", "S", "SW", "W", "NW", "N"]
    return pd.cut(derajat_series, bins=bins, labels=labels, right=False, ordered=False)

# Menerapkan pd.cut() ke kolom 'derajat' dan membuat kolom baru 'arah'
df['DDD_X_CAR'] = derajat_ke_arah_pandas(df['DDD_X'])

print(df)
```

Kemungkinan, Null pada kolom DDD_X_CAR diakibatkan oleh nilai 360 yang tidak terisi oleh 'N'

```python
# Mengisi nilai kosong dengan 'N'
df['DDD_X_CAR'] = df['DDD_X_CAR'].fillna('N')
```

```python
def buat_mawar_angin(arah, judul):
    """Membuat diagram mawar angin."""

    arah_rad = np.radians(arah)

    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='polar')

    ax.hist(arah_rad, bins=8, alpha=0.7)  # Gunakan histogram
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.set_title(judul, y=1.1)

    plt.show()
```

```python
# Kelompokkan data berdasarkan bulan
df['Bulan'] = df['TANGGAL'].dt.month_name()

for bulan, data_bulan in df.groupby('Bulan'):
    buat_mawar_angin(data_bulan['DDD_X'], f'Diagram Arah Angin - {bulan}')
```
